```python
import pandas as pd

train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')
```

```python
x_train = train.drop(columns=['ID', '이용금액'])
y_train = train['이용금액']
x_test = test.drop(columns=['ID'])
```

```python
x_train_dummy = pd.get_dummies(x_train)
x_test_dummy = pd.get_dummies(x_test)[x_train_dummy.columns]
```

```python
from sklearn.model_selection import train_test_split

X_t, X_v, y_t, y_v = train_test_split(x_train_dummy, y_train, random_state=42, test_size=0.33)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(random_state=23)
rfr.fit(X_t, y_t)
```

```python
ptl = rfr.predict(X_t)
pvl = rfr.predict(X_v)
p_test_l = rfr.predict(x_test_dummy)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

print(np.sqrt(mean_squared_error(y_t, ptl)))
print(np.sqrt(mean_squared_error(y_v, pvl)))
```

```python
test['이용금액'] = p_test_l

test[['ID', '이용금액']]
```
