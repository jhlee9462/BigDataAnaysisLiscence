```python
import pandas as pd

x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/X_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/shipping/y_test.csv')
```

```python
drop_columns = ['ID']

x_train_drop = x_train.drop(columns=drop_columns)
x_test_drop = x_test.drop(columns=drop_columns)
```

```python
y_train_t = y_train.iloc[:,1]
y_test_t = y_test.iloc[:,1]
```

```python
x_train_dummy = pd.get_dummies(x_train_drop)
x_test_dummy = pd.get_dummies(x_test_drop)[x_train_dummy.columns]
```

```python
from sklearn.model_selection import train_test_split

X_t, X_v, y_t, y_v = train_test_split(x_train_dummy, y_train_t, test_size=0.33, random_state=42, stratify=y_train_t)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=23)
rfc.fit(X_t, y_t)
```

```python
from sklearn.metrics import f1_score

ptl = rfc.predict(X_t)
pvl = rfc.predict(X_v)
p_test_l = rfc.predict(x_test_dummy)

print(f'{f1_score(y_t, ptl)=}')
print(f'{f1_score(y_v, pvl)=}')
print(f'{f1_score(y_test_t, p_test_l)=}')
```
