```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/normal1.csv')
plt.hist(df)
plt.show()
```

```python
from scipy.stats import shapiro
print(shapiro(df))
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/normal3.csv')
plt.hist(df2)
plt.show()
```

```python
print(shapiro(df2))
```

```python
import numpy as np

df3 = np.log1p(df2)
plt.hist(df3)
plt.show()
```

```python
print(shapiro(df3))
```
