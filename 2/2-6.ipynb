```python
import pandas as pd

x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/y_test.csv')
```

```python
# 결측치 대체
pm = x_train['ph'].mean()
sm = x_train['Sulfate'].mean()
tm = x_train['Trihalomethanes'].mean()

x_train['ph'] = x_train['ph'].fillna(pm)
x_train['Sulfate'] = x_train['Sulfate'].fillna(sm)
x_train['Trihalomethanes'] = x_train['Trihalomethanes'].fillna(tm)
x_test['ph'] = x_test['ph'].fillna(pm)
x_test['Sulfate'] = x_test['Sulfate'].fillna(sm)
x_test['Trihalomethanes'] = x_test['Trihalomethanes'].fillna(tm)
```

```python
# 컬럼 제거
drop_columns = ['ID']
x_train_drop = x_train.drop(columns=drop_columns)
x_test_drop = x_test.drop(columns=drop_columns)[x_train_drop.columns]
```

```python
# y값 추출
y_train_t = y_train['Potability']
y_test_t = y_test['Potability']
```

```python
from sklearn.model_selection import train_test_split

X_t, X_v, y_t, y_v = train_test_split(x_train_drop, y_train_t, test_size=0.33, random_state=42, stratify=y_train_t)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=23)
rfc.fit(X_t, y_t)
```

```python
ptl = rfc.predict(X_t)
ptp = rfc.predict_proba(X_t)[:, 1]
pvl = rfc.predict(X_v)
pvp = rfc.predict_proba(X_v)[:, 1]
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_auc_score

print(f'train accuracy score: {accuracy_score(y_t, ptl)}')
print(f'train f1 score: {f1_score(y_t, ptl)}')
print(f'train recall score: {recall_score(y_t, ptl)}')
print(f'train precision score: {precision_score(y_t, ptl)}')
print(f'train roc auc score: {roc_auc_score(y_t, ptp)}\n')

print(f'validation accuracy score: {accuracy_score(y_v, pvl)}')
print(f'validation f1 score: {f1_score(y_v, pvl)}')
print(f'validation recall score: {recall_score(y_v, pvl)}')
print(f'validation precision score: {precision_score(y_v, pvl)}')
print(f'validation roc auc score: {roc_auc_score(y_v, pvp)}')
```

```python
p_test_l = rfc.predict(x_test_drop)
p_test_p = rfc.predict_proba(x_test_drop)[:, 1]

print(f'test accuracy score: {accuracy_score(y_test_t, p_test_l)}')
print(f'test f1 score: {f1_score(y_test_t, p_test_l)}')
print(f'test recall score: {recall_score(y_test_t, p_test_l)}')
print(f'test precision score: {precision_score(y_test_t, p_test_l)}')
print(f'test roc auc score: {roc_auc_score(y_test_t, p_test_p)}')
```
