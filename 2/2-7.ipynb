```python
import pandas as pd

x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/drug/y_test.csv')
```

```python
drop_columns = ['ID']
x_train_drop = x_train.drop(columns=drop_columns)
x_test_drop = x_test.drop(columns=drop_columns)
```

```python
x_train_d = pd.get_dummies(x_train_drop)
x_test_d = pd.get_dummies(x_test_drop)[x_train_d.columns]
```

```python
y_train_t = y_train['Drug']
y_test_t = y_test['Drug']
```

```python
from sklearn.model_selection import train_test_split

X_t, X_v, y_t, y_v = train_test_split(x_train_d, y_train_t, test_size=0.33, random_state=42, stratify=y_train_t)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=23)
rfc.fit(X_t, y_t)
```

```python
ptl = rfc.predict(X_t)
ptp = rfc.predict_proba(X_t)[:, 1]
pvl = rfc.predict(X_v)
pvp = rfc.predict_proba(X_v)[:, 1]
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(f'train accuracy score : {accuracy_score(y_t, ptl)}')
print(f'train f1 score : {classification_report(y_t, ptl)}')

print(f'validation accuracy score : {accuracy_score(y_v, pvl)}')
print(f'validation f1 score : {classification_report(y_v, pvl)}')
```

```python
p_test_l = rfc.predict(x_test_d)
p_test_p = rfc.predict_proba(x_test_d)
```
