```python
import pandas as pd
video = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
channel = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
```

```python
video.ct = pd.to_datetime(video.ct)
channel.ct = pd.to_datetime(channel.ct)
video.value_counts('videoname')
```

```python
video.sort_values(['videoname', 'ct']).drop_duplicates('videoname', keep='last')[['viewcnt', 'videoname', 'ct']].reset_index(drop=True)
```

```python
channel.loc[channel.ct >= pd.to_datetime('2021-10-03')].sort_values(['channelid', 'ct']).drop_duplicates('channelid')[['channelname', 'subcnt']].reset_index(drop=True)
```

```python
import datetime

video.videoname = video.videoname.astype('string')
ep1 = video.loc[video.videoname.str.contains('EP1')].sort_values('ct').reset_index(drop=True)
target = ep1.loc[
    (ep1.ct.diff(1) >= datetime.timedelta(minutes=20)) | \
    (ep1.ct.diff(1) <= datetime.timedelta(minutes=5))
]

ep1.loc[ep1.index.isin([720,721,722,723,1635,1636,1637])]
```

```python
target = video.sort_values(['ct', 'videoname']).drop_duplicates('videoname').copy()
target['date'] = target.ct.dt.date
target[['date', 'videoname']]
```

```python
video['time'] = video.ct.dt.hour
video.loc[video.time == 21][['videoname', 'viewcnt', 'ct']].sort_values('ct').drop_duplicates('videoname').sort_values('viewcnt', ascending=False).reset_index(drop=True)
```

```python
target = video[['videoname', 'ct', 'dislikecnt', 'likecnt']].sort_values(['videoname', 'ct'], ascending=False).drop_duplicates('videoname')
target['ratio'] = target.apply(lambda x: x.dislikecnt / x.likecnt, axis=1)
target[['videoname', 'ratio']].sort_values('ratio').reset_index(drop=True)
```

```python
target = video.loc[
    (video.ct >= pd.to_datetime('2021-11-01 00:00:00')) & \
    (video.ct <= pd.to_datetime('2021-11-01 15:00:00'))
].reset_index(drop=True)

target[['videoname', 'viewcnt']].groupby('videoname').agg(lambda x: max(x) - min(x))
```
