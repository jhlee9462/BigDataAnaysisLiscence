```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/kingcountyprice/train.csv')
df.head()
```

```python
mpb = df.groupby('bedrooms').size().sort_values(ascending=False).index[0]

t = df.loc[df['bedrooms'] == mpb]['price']
print(t.quantile(0.9) - t.quantile(0.1))
```
