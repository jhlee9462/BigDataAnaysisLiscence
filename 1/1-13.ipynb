```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/mobile/train.csv')
df.head()
```

```python
df[['price_range', 'n_cores']].groupby(['price_range', 'n_cores']).size().sort_values(ascending=False).reset_index().groupby('price_range').head(1)
```

```python
target = df.loc[df['price_range'] == 3].corr().unstack().sort_values(ascending=False).reset_index().rename(columns={0: 'corr'})
target.loc[target['corr'] != 1].iloc[1]
```
