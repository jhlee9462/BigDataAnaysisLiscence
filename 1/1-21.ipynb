```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/train.csv')
df.head()
```

```python
target = df.groupby('pose').median()
result = abs(target.iloc[0] - target.iloc[1])
result = result.sort_values(ascending=False).rename('diff').to_frame()
list(result.loc[result['diff'] == result.iloc[0].max()].index)
```
